### Variables de Interés

Nuestras variables de interés son:
- GDP (in Trillions USD)
- GDP per Capita (in USD)
- Life Expectancy (Years)
- Education Expenditure
- Literacy Rate (%)
- Healthcare Expenditure per Capita (USD)
- Human Development Index (HDI)
- Crime Rate (per 100,000)
- Unemployment Rate (%)
- Region
- Country
- Continent

```python
# PreProcesamiento para filtrar datos no deseados
# Lista de las columnas que nos interesan
columns_of_interest = [
    'Country',
    'Year',
    'GDP (in Trillions USD)',
    'GDP per Capita (in USD)',
    'Life Expectancy (Years)',
    'Healthcare Expenditure per Capita (USD)',
    'Crime Rate (per 100,000)',
    'Unemployment Rate (%)',
    'Literacy Rate (%)',
    'Human Development Index (HDI)'
]

# Filtrar las columnas de interés
db_cleaned = db[columns_of_interest]

# Mostrar los primeros datos de la base ya limpia
print("Datos limpios:")
print(db_cleaned.head())
```

```python
# Histogramas y diagramas de cajas para visualizar distribuciones en variables de interés

# Lista de variables numéricas de interés para graficar
variables_numericas = [
    'GDP (in Trillions USD)',
    'GDP per Capita (in USD)',
    'Life Expectancy (Years)',
    'Healthcare Expenditure per Capita (USD)',
    'Crime Rate (per 100,000)',
    'Unemployment Rate (%)'
]

# Graficar histogramas y diagramas de cajas para cada variable de interés
for var in variables_numericas:
    plt.figure(figsize=(14, 6))

    # Histograma
    plt.subplot(1, 2, 1)
    sns.histplot(db_cleaned[var], bins=20, kde=True)
    plt.title(f"Histograma de {var}")

    # Diagrama de caja
    plt.subplot(1, 2, 2)
    sns.boxplot(x=db_cleaned[var])
    plt.title(f"Diagrama de caja de {var}")

    # Mostrar las gráficas
    plt.tight_layout()
    plt.show()
```

```python
# Media y desviación estándar de las variables de interés
for var in variables_numericas:
    media = db_cleaned[var].mean()
    desviacion_estandar = db_cleaned[var].std()
    print(f"{var} \n- Media: {media:.2f}, Desviación estándar: {desviacion_estandar:.2f} \n")
```

```python
#Grafica de dispersion de variables de interes de una a una
import itertools

db_cleaned = db_cleaned.dropna()
# combinaciones de pares de variables de una a una
combinations = list(itertools.combinations(db_cleaned, 2))

for x_var, y_var in combinations:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=db_cleaned, x=x_var, y=y_var, hue='Country', palette='tab10', alpha=0.5)#coloreado por pais

    plt.xlabel(x_var)
    plt.ylabel(y_var)
    plt.title(f'{y_var} en función de {x_var}')
    plt.show()
```

```python
# Correlación entre las variables de interés

# Calcular la matriz de correlación
correlation_matrix = db_cleaned[variables_numericas].corr()

# Visualizar la matriz de correlación con un heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Matriz de correlación entre variables de interés")
plt.show()
```
