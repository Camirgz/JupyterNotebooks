### 5.  Para la variable de esperanza de vida, calcule la media, mediana y la desviación estándar de todos los países, para el año más reciente. Imprima estos valores en el notebook.

```python
# Usando df_life_expectancy_no_null y asignarlo a df_recent_clean, para que queden los valores requeridos en los años más recientes.
df_recent_clean = df_life_expectancy_no_null.loc[df_life_expectancy_no_null.groupby('Country')['Year'].idxmax()]

# Calcular las estadísticas de la columna 'Life expectancy'.
mean_life_expectancy = df_recent_clean['Life expectancy '].mean()
median_life_expectancy = df_recent_clean['Life expectancy '].median()
std_life_expectancy = df_recent_clean['Life expectancy '].std()

# Imprimir los resultados.
print("Media de esperanza de vida: ", mean_life_expectancy)
print("Mediana de esperanza de vida: ", median_life_expectancy)
print("Desviación estándar de esperanza de vida: ", std_life_expectancy)
```

### 6. Realice un histograma con la variable esperanza de vida para conocer cómo se distribuyen los datos. Cambie progresivamente el número de barras, desde 5 hasta 100. Reporte qué cambia en las distribuciones conforme aumenta la cantidad de barras. Escoja qué número de barras le parece óptimo y explique por qué. Investigue en alguna fuente fiable qué criterios se pueden utilizar y reporte un enlace a esta fuente.

```python
# Generar histograma para diferentes números de barras (bins)
bins_list = [5, 10, 20, 50, 100]

# Crear una figura con subplots para mostrar todos los histogramas
fig, axs = plt.subplots(len(bins_list), figsize=(10, 15))
fig.tight_layout(pad=5.0)

# Dibujar histogramas con diferentes números de barras
for i, bins in enumerate(bins_list):
    axs[i].hist(df_recent_clean['Life expectancy '], bins=bins, edgecolor='black')
    axs[i].set_title(f'Histograma con {bins} barras')
    axs[i].set_xlabel('Esperanza de vida')
    axs[i].set_ylabel('Frecuencia')

# Mostrar la figura
plt.show()
```

Conforme aumenta la cantidad de barras en los histogramas, la representación de la distribución de datos se vuelve más específica. Con pocas barras (5 o 10), el histograma muestra tendencias generales y picos grabder, como se ve alrededor de 75. Con más barras (20, 50, 100), se revela mayor variabilidad y detalles en la distribución, pero también puede aparecer ruido datos no relevantes.
Un número óptimo de barras podría ser 10 o 20, ya que equilibra el detalle con la claridad. Esto está en línea con la regla de Sturges, que sugiere un número de barras basado en el tamaño de la muestra para evitar demasiada simplicidad o más bien demasiado detalle.

### 7. En la misma gráfica anterior, coloque líneas verticales en la media (μ) y en la posiciones correspondientes a μ más una desviación estándar (σ) y menos una desviación estándar: μ - σ y μ + σ. Comente qué particularidades tiene esta región alrededor de la media.

```python
# Ya obtuvimos previamente la media (μ) y la desviación estándar (σ)

# Generar histogramas para diferentes números de barras (bins)
bins_list = [5, 10, 20, 50, 100]

# Crear una figura con subplots para mostrar todos los histogramas
fig, axs = plt.subplots(len(bins_list), figsize=(10, 15))
fig.tight_layout(pad=5.0)

# Dibujar histogramas con diferentes números de barras y añadir líneas verticales
for i, bins in enumerate(bins_list):
    axs[i].hist(df_recent_clean['Life expectancy '], bins=bins, edgecolor='black')

    # Añadir líneas verticales en la media (μ), μ - σ y μ + σ
    axs[i].axvline(mean_life_expectancy, color='red', linestyle='dashed', linewidth=1, label=f'μ = {mean_life_expectancy:.2f}')
    axs[i].axvline(mean_life_expectancy - std_life_expectancy, color='blue', linestyle='dashed', linewidth=1, label=f'μ - σ = {mean_life_expectancy - std_life_expectancy:.2f}')
    axs[i].axvline(mean_life_expectancy + std_life_expectancy, color='blue', linestyle='dashed', linewidth=1, label=f'μ + σ = {mean_life_expectancy + std_life_expectancy:.2f}')

    axs[i].set_title(f'Histograma con {bins} barras')
    axs[i].set_xlabel('Esperanza de vida')
    axs[i].set_ylabel('Frecuencia')
    axs[i].legend()

# Mostrar la figura
plt.show()
```

Podemos observar que la región se encuentra marcada por las líneas punteadas azules. Primero que todo, podemos observar que la mayoría de los valores de la esperanza de vida se encuentran  dentro del rango de μ ± σ (entre 63.49 y 79.74). Esto refleja el principio general de las distribuciones normales, donde aproximadamente el 68% de los datos se encuentran en este rango.
Conforme aumenta el número de barras, este patrón se mantiene, con una mayor densidad de datos concentrada alrededor de la media.
Asimismo, en este caso, los datos parecen bastante simétricos alrededor de la media (71.62). Esto indica que la distribución de la esperanza de vida tiene una forma cercana a una campana, lo cual es típico de una distribución normal.
A medida que nos alejamos de la media, la frecuencia de los valores disminuye. Esto es más notable cuando se usan más barras, como en el histograma con 100 barras, donde se puede ver una caída más gradual fuera de la región de la media y más hacia los extremos.

### 11. Usando la base de datos world-cities-database, y algún sitio o servicio web (o biblioteca de Python), cree una función que permita obtener la altitud correspondiente a una ciudad, utilizando la latitud y longitud.

```python
import requests

# Cargar la base de datos
datos_ciudades = pd.read_csv('worldcitiespop.csv', low_memory=False)

# Función para obtener la elevación a partir del nombre de una ciudad
def obtener_elevacion_por_nombre(ciudad):
    # Buscar la ciudad
    info_ciudad = datos_ciudades[datos_ciudades['City'].str.lower() == ciudad.lower()]

    if not info_ciudad.empty:
        # Obtener latitud y longitud
        lat = info_ciudad['Latitude'].values[0]
        lon = info_ciudad['Longitude'].values[0]

        # Llamar a la API para obtener la elevación
        api_url = "https://api.open-elevation.com/api/v1/lookup"
        parametros = {'locations': f"{lat},{lon}"}
        respuesta = requests.get(api_url, params=parametros)

        if respuesta.status_code == 200:
            contenido = respuesta.json()
            elevacion = contenido['results'][0]['elevation']
            return lat, lon, elevacion
        else:
            print("Intenta de nuevo.")
            return lat, lon, None

    return None, None, None

# Ejemplo de uso de la función
nombre = "aixas"  # Cambia el nombre de la ciudad por la que desees consultar
lat, lon, elevacion_ciudad = obtener_elevacion_por_nombre(nombre)

# Imprimir los resultados obtenidos
if elevacion_ciudad is not None:
    print(f"Ciudad: {nombre}, Latitud: {lat}, Longitud: {lon}, Elevación: {elevacion_ciudad} metros")
else:
    print(f"No se encontró información para la ciudad: {nombre}")

```

### 12. Para las ciudades con población mayor a 100,000 habitantes, obtenga la altitud correspondiente y agregue todos estos valores en una nueva columna de la base de datos.

```python
import pandas as pd
import requests

# Filtrar las ciudades con población mayor a 100,000
ciudades_grandes = datos_ciudades[datos_ciudades['Population'] > 100000].copy()

# Crear una nueva columna para la elevación
ciudades_grandes.loc[:, 'Elevation'] = None
#Obtener la elevación
for index, row in ciudades_grandes.iterrows():
    ciudad = row['City']
    elevacion = obtener_elevacion_por_nombre(ciudad)
    ciudades_grandes.at[index, 'Elevation'] = elevacion

# Agregar la nueva columna al original
datos_ciudades = datos_ciudades.merge(ciudades_grandes[['City', 'Elevation']], on='City', how='left')
```
